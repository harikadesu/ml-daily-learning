```python
# Day 1 - Machine Learning Basics + DSA
# Author: Harika

# Section 1: Supervised Learning
print("Supervised Learning Example: Spam Detection")

from sklearn.linear_model import LogisticRegression

X = [[0, 1], [1, 0], [1, 1], [0, 0]]  # Email features (word presence)
y = [1, 0, 1, 0]  # 1 = Spam, 0 = Not Spam

model = LogisticRegression()
model.fit(X, y)
print("Prediction for [1, 0]:", model.predict([[1, 0]]))

# Section 2: Unsupervised Learning
print("\nUnsupervised Learning Example: Customer Segmentation")

from sklearn.cluster import KMeans

X = [[25, 5], [40, 15], [20, 4], [42, 16]]  # [Age, Purchases/month]

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
print("Cluster labels:", kmeans.labels_)

# Section 3: Arrays in Python (DSA Basics)
print("\nArray Example: Weekly Temperature Data")

temperatures = [74, 75, 73, 70, 72, 76, 78]
print("Monday’s temperature:", temperatures[0])
print("Average temperature:", sum(temperatures)/len(temperatures))
```
